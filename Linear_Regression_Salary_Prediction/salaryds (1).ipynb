```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
saldf = pd.read_csv('/content/Salary Data.csv')
saldf.head()
```

```python
saldf.info()
```

```python
saldf.isnull().sum()
```

```python
inp = saldf[['Years of Experience']]
out = saldf['Salary']
```

```python
LR = LinearRegression()
```

```python
df_cleaned = saldf.dropna(subset=['Years of Experience', 'Salary'])

inp_cleaned = df_cleaned[['Years of Experience']]
out_cleaned = df_cleaned['Salary']

LR.fit(inp_cleaned, out_cleaned)
```

```python
LR.predict([[5]])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(inp_cleaned[:1000], out_cleaned[:1000])  # plot only first 1000 for speed
plt.plot(inp_cleaned[:1000], LR.predict(inp_cleaned[:1000]), linewidth=3)
plt.title("Linear Regression: Years of Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
