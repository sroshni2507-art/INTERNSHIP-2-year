```python
# ==============================
# IMPORT LIBRARIES
# ==============================
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# LOAD PLANETS DATASET
# ==============================
df = sns.load_dataset("planets")

# ==============================
# BASIC INFO
# ==============================
print("Dataset Shape:", df.shape)
print("\nColumn Names:")
print(df.columns)

print("\nFirst 5 Rows:")
df.head()
```

```python
plt.figure(figsize=(10,5))
df["year"].value_counts().sort_index().plot(kind="line")
plt.xlabel("Year")
plt.ylabel("Number of Planets Discovered")
plt.title("Planets Discovered Per Year")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="method")
plt.xticks(rotation=45)
plt.title("Planet Discovery Methods")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x="orbital_period",
    y="mass",
    hue="method"
)
plt.title("Orbital Period vs Planet Mass")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x="orbital_period",
    y="mass",
    hue="method"
)
plt.title("Orbital Period vs Planet Mass")
plt.show()
```

```python
pivot = pd.pivot_table(
    df,
    values="number",
    index="year",
    columns="method",
    aggfunc="count"
)

plt.figure(figsize=(12,6))
sns.heatmap(pivot.fillna(0), cmap="Blues")
plt.title("Planet Discoveries by Year and Method")
plt.show()
```
