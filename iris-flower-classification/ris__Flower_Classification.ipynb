```python
import pickle
from google.colab import files

df = sns.load_dataset("iris") # Added to resolve NameError

# ðŸŽ¯ FEATURES & TARGET
X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y = df["species"]

# âœ‚ SPLIT INTO TRAIN AND TEST
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Save the split data to a .pkl file for download
data_to_save = {
    'X_train': X_train,
    'X_test': X_test,
    'y_train': y_train,
    'y_test': y_test
}
file_name = 'iris_split_data.pkl'
with open(file_name, 'wb') as f:
    pickle.dump(data_to_save, f)

print(f"'{file_name}' has been created.")
files.download(file_name)
```

```python
dt = DecisionTreeClassifier(max_depth=4, random_state=42)
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
# ðŸ“Œ VISUALIZE THE DECISION TREE
plt.figure(figsize=(15,8))
plot_tree(dt, feature_names=X.columns, class_names=dt.classes_, filled=True)
plt.show()
```
